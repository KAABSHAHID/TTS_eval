### Data Prep
We now use the `Qwen-3` format for conversation style finetunes. We use [Maxime Labonne's FineTome-100k](https://huggingface.co/datasets/mlabonne/FineTome-100k) dataset in ShareGPT style. Qwen-3 renders multi turn conversations like below:

```
<|im_start|>user
Hello!<|im_end|>
<|im_start|>assistant
Hey there!<|im_end|>

```
We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3, phi4, qwen2.5, gemma3` and more.

````python
from unsloth.chat_templates import get_chat_template
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "qwen3-instruct",
)
````

````python
import pandas as pd
import pandas as pd
from datasets import Dataset
# Load your CSV
data = pd.read_csv("sarcasm_cleaned.csv", encoding="utf-8")
df = data[-1000:]
print(len(df))

# Create a new column with the conversation structure
def make_conversation(row):
    return {
        "conversations": [
            {"content": row["text"], "role": "user"},     # prompt + text
            {"content": row["label"], "role": "assistant"} # sarcasm label
        ]
    }

df["conversations"] = df.apply(make_conversation, axis=1)

# Keep only the 'conversations' column
df = df[["conversations"]]

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(df)

# Now you can access like dataset[100]
print(dataset[100])
````

We now have to apply the chat template for `Qwen-3` onto the conversations, and save it to `text`.

````python
def formatting_prompts_func(examples):
   convos = examples["conversations"]
   texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
   return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)
````

````python
def formatting_prompts_func(example):
    # unwrap if nested
    convo = example["conversations"]
    if isinstance(convo, dict) and "conversations" in convo:
        convo = convo["conversations"]

    text = tokenizer.apply_chat_template(
        convo,
        tokenize=False,
        add_generation_prompt=False
    )
    return {"text": text}

dataset = dataset.map(formatting_prompts_func, batched=False)
````

### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

````python
from trl import SFTTrainer, SFTConfig
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None, # Can set up evaluation!
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4, # Use GA to mimic batch size!
        warmup_steps = 5,
        num_train_epochs = 1, # Set this for 1 full training run.
        #max_steps = 60,
        learning_rate = 2e-4, # Reduce to 2e-5 for long training runs
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none", # Use this for WandB etc
    ),
)
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs. This helps increase accuracy of finetunes!

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|im_start|>user\n",
    response_part = "<|im_start|>assistant\n",
)
````

````python
data = pd.read_csv("sarcasm_cleaned.csv", encoding="utf-8")
data = data[-1000:]
dataset = Dataset.from_pandas(data)

dataset
````

````python
import torch
from tqdm import tqdm

pred_labels = []

for example in tqdm(dataset):
    text = example["text"]
    messages = [{"role": "user", "content": text}]
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=3,  # only €Åÿß⁄∫ / ŸÜ€Å€å⁄∫
            temperature=0.0,
            do_sample=False
        )

    pred = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    pred_labels.append(pred)

# Add predictions to dataset
dataset = dataset.add_column("pred_label", pred_labels)
````

````python
dataset["pred_label"]
````

````python
clean_preds = []

for pred in pred_labels:
    # Split by '\nassistant' and take last part
    if "\nassistant" in pred:
        label = pred.split("\nassistant")[-1].strip()
    else:
        label = pred.strip()
    # Optional: normalize common typos
    if label in ["ŸÜ€Å€å", "ŸÜ€Å€å⁄∫", "ŸÜ€ÅŸä"]:
        label = "ŸÜ€Å€å⁄∫"
    elif label in ["€Åÿß⁄∫", "€Åÿß", "Ÿáÿß⁄∫"]:
        label = "€Åÿß⁄∫"
    clean_preds.append(label)

# Replace old pred_labels with cleaned ones
dataset = dataset.add_column("pred_label_clean", clean_preds)
````

````python
dataset["pred_label_clean"]
````

````python
correct = sum(dataset[i]["label"] == dataset[i]["pred_label_clean"] for i in range(len(dataset)))
accuracy = correct / len(dataset)
print(f"Accuracy: {accuracy*100:.2f}%")
````

````python
tp = sum((dataset[i]["label"] == "€Åÿß⁄∫") and (dataset[i]["pred_label_clean"] == "€Åÿß⁄∫") for i in range(len(dataset)))
fp = sum((dataset[i]["label"] == "ŸÜ€Å€å⁄∫") and (dataset[i]["pred_label_clean"] == "€Åÿß⁄∫") for i in range(len(dataset)))
fn = sum((dataset[i]["label"] == "€Åÿß⁄∫") and (dataset[i]["pred_label_clean"] == "ŸÜ€Å€å⁄∫") for i in range(len(dataset)))

precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0
f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

print(f"F1-score (€Åÿß⁄∫ as positive): {f1:.4f}")
````

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.
